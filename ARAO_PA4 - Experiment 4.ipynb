```python
# Access the libraries

# Used for numerical computations 
import numpy as np

# Used for data structure manipulations and operations 
import pandas as pd

# Used for visualization
import matplotlib.pyplot as plt
```

#### PROBLEM 1

```python
# Read the excel file

boards = pd.read_excel('board2.xlsx')
boards
```

##### A. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
# Create a DataFrame called 'Instru' by identifying the constants and showing the entries where the `Name`, `GEAS`, and `Electronics` grades are all above 70.

Instru = pd.DataFrame(boards.loc[(boards['Hometown']=='Luzon')&(boards['Electronics']>70)&(boards['Track']=='Instrumentation'), ['Name', 'GEAS', 'Electronics']])
Instru
```

##### B. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Femalee

```python
# Create a DataFrame named 'Mindy' by identifying the constants and showing the `Name`, `Track`, `Electronics`, and `Average` grades where each grade is greater than or equal to 55.

boards['Average'] = round(boards[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1),2)
Mindy = boards.loc[(boards['Hometown'] == 'Mindanao') & (boards['Gender'] == 'Female') & (boards['Average'] >= 55), ['Name','Track','Electronics', 'Average']]
Mindy
```

#### PROBLEM 2

```python
plt.figure (figsize = (5,4))
plt.bar(boards["Track"], boards["Average"])
plt.ylabel ("Track", fontsize = 8)
plt.xlabel ("Average", fontsize = 8)
plt.title ("ECE Track vs. Average", fontsize = 10)
plt.show()
```

##### The bar graph reveals that the Microelectronics Track has the highest average grades. On the other hand, the Instrumentation Track has the lowest average grades.
##### --------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
plt.figure (figsize = (5,4))
plt.bar(boards["Hometown"], boards["Average"])
plt.ylabel ("Hometown", fontsize = 8)
plt.xlabel ("Average", fontsize = 8)
plt.title ("Hometown vs. Average", fontsize = 10)
plt.show()
```

##### The bar graph shows that Luzon generally outperformed those from Mindanao and Visayas.

```python
plt.figure (figsize = (5,4))
plt.bar(boards["Gender"], boards["Average"])
plt.ylabel ("Gender", fontsize = 8)
plt.xlabel ("Average", fontsize = 8)
plt.title ("Gender vs. Average", fontsize = 10)
plt.show()
```

##### The bar graph shows that female examinees have the highest average grade than male examinees.
